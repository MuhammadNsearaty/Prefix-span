## getFreq (paths)
- this function takes a list:paths of the target data as list of strings
- it return's a Dictionary with the frequent of each item 

```python
def getFreq(paths):
    """get The 1st item Freq"""
    freqDict ={}
    for item in paths:
        currPaths = item.split(';')
        for token in currPaths:
            if(token not in freqDict):
                freqDict[token] = 1
            else:
                freqDict[token] +=1
    return freqDict
```

## checkMinSup(minSup,freqs)
- it takes the minimum limit of the allowed minSupport and the current patterns with thier frequent
- return's the allowed patterns in Dictionary

```python
def checkMinSup(minSup,freqs):
    """Check MinSupport for each seq"""
    res = {}
    for key in freqs.keys():
        if freqs[key] >= minSup:
            res[key] = freqs[key]
    return res
```

## getSeqFreq(seqList,paths)
- this function takes the seqList (the currPatterns list ) and the paths ( the currDataBase )
- return's the frequent of each pattern in Dictionary

```python
def getSeqFreq(seqList,paths):
    """get the seq freq for the k lvl"""
    resDict = {}
    for seq in seqList:
#         splitSeq = seq.split(',')
        freq = 0

        for path in paths:
            res = True
            for item in seq:
                if item not in path:
                    res = False
                    break
                
            if res == True:
                freq+=1
        resDict[','.join(seq)] = freq
            
    return resDict
```

## alterDB(dataset,a)

```python
def alterDB(dataset,a):
    newDataset =[]
    for path in dataset:     
        items = path.split(';')
        newPath = []
        found = False
        for i in range(len(items)):      
            if a == items[i]:
                found = True
                if i == len(items)-1:
                    newPath=[]
                else:
                    newPath = items[i+1:]
                break
        if found==False:
            newPath = []
        found = False
        if newPath:
            newDataset.append(';'.join(newPath))
    return newDataset
```

## prefix_span(currSeq,subSeq,alpha)
- it takes the current suquence and pass it down the Tree
- the subSeq is the current Database or in another word it's the Projected DataBase of the currSeq of the prefix of the currSeq
- alpha is a list contains all finished patterns
- NOTE: remove the comment from the Print line to see the stages of the search ( not recommended :( )  

```python
def prefix_span(currSeq,subSeq,alpha):
    if not subSeq:
        alpha.append(currSeq)
        return alpha
    freqItems = getFreq(paths=list(subSeq))
    freqItems = checkMinSup(minSup,freqItems)
    res =[]
    for item in list(freqItems.keys()):
        """ex: <a,c,b,f> for each a,c... item in the currPath"""
        newData = alterDB(subSeq,item)
        newSeq = currSeq.copy()
        newSeq.append(item)
#         print(f'The Current DataBase{subSeq} \n the Current Sequence{currSeq} \n New Sequence {newSeq}\n New DataBase{newData}\nCurrent alpha{alpha}\n--------------------\n')
        alpha = prefix_span(newSeq,newData,alpha)
       
    return alpha
    
        
```
